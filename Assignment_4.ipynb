# Using imodels models on the 'Wine Quality' Dataset



```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('winequality-red.csv', sep = ';')

print(data.head)
```

###This dataset has 12 features. What I would like to do is use the previous 11 to predict the 'quality' metric.

```python
from imodels import RuleFitRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mse
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

## RuleFit Regressor

When I trained the model on the raw data it tended to guess 5 or 6 evenery time, this is because the vast majority of the wines are listed at that quality, we have to scale the data to force the model to learn the different qualities. I chose to oversample as the minority class is important and i didn't want to lose information by undersampling

```python
X = data.drop('quality', axis = 1)
y = data['quality']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

smote = SMOTE(random_state = 0)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

#scaler = StandardScaler()
#X_train_scaled = scaler.fit_transform(X_train_balanced) #I took out scaling because interpretability suffered more than accuracy improved
#X_test_scaled = scaler.transform(X_test)

model = RuleFitRegressor(random_state=0)
model.fit(X_train_balanced, y_train_balanced)

y_pred = model.predict(X_test)

MeanSE = mse(y_pred, y_test)

print("Model mean squared error:", MeanSE)
beta_0 = model.intercept
print("Beta 0 is",beta_0)

model.visualize()
```

##GreedyTree Regressor

```python
from imodels import GreedyTreeRegressor
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

smote = SMOTE(random_state = 0)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

model = GreedyTreeRegressor(random_state = 0, max_depth=4) #Sacrifice a little accuracy fro interpretability/visualisation
model.fit(X_train_balanced, y_train_balanced)

y_pred = model.predict(X_test)
MeanSE2 = mse(y_pred, y_test)

print(MeanSE2)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (128,96))
plot_tree(model, filled=True, feature_names=X.columns)
plt.show()
```

## Boosted Rules Regressor

```python
from imodels import BoostedRulesRegressor
from sklearn.tree import export_text
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

smote = SMOTE(random_state = 0)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

model = BoostedRulesRegressor(random_state = 0)
model.fit(X_train_balanced, y_train_balanced)

y_pred = model.predict(X_test)
MeanSE2 = mse(y_pred, y_test)

importances = model.feature_importances_

print("Model mse:", MeanSE2)
print("According to this model, quality is determined by:")
for i, feature in enumerate(X.columns):
  print(feature, "contributes", str(importances[i] *100) +"%")
```
